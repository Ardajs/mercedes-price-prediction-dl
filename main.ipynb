```python
dataFrame = clean_data_Df
```

```python
dataFrame = dataFrame[dataFrame["year"] != 1970 ]
```

```python
dataFrame = dataFrame.drop("transmission", axis=1)
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))



model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(x_train_scaled, y_train, validation_data=(x_test_scaled, y_test), batch_size=250, epochs=300)
```

```python
lossData = pd.DataFrame(model.history.history)
```

```python
lossData.head()
```

```python
lossData.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
guessArray = model.predict(x_test_scaled)
guessArray
```

```python
mean_absolute_error(y_test, guessArray)
```

```python
plt.scatter(x=y_test, y=guessArray)
plt.plot(y_test, y_test, "g--")
```
